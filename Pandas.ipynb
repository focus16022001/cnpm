Tệp CSV

```python
#path_or_buf: đối số đầu tiên của CSV => đường dẫn tệp 
# nếu bỏ qua path_or_buf: .to_csv() không tạo bất kì tệp nào => trả về chuổi tương ứng
df = pd.DataFrame(data=data).T
s = df.to_csv()
print(s)
```

```python
#pandas sử dụng giá trị NaN để thay thế các giá trị bị thiếu
# .loc[] lấy dữ liệu trên hàng và cột
df.loc['RUS', 'CONT']
```

```python
#KHI LUWU DataFrame vào CSV, chuỗi trống ('') đại diện cho dữ liệu bị thiếu
# nếu thay đổi hành vi chọn na_rep
# khi đọc tệp một vài chỗ khác bị thiếu: nan, -nan, NA, N/A, NaN, null
df.to_csv('newdata.csv', na_rep = '(missing)')
s = pd.read_csv('newdata.csv', index_col=0)
s
# có thể sử dụng read_csv(): 
#pd.read('newdata.csv', index_col=0, na_values='(missing)')
```

```python
# kiểm tra dữ liệu cho các cột
df.dtypes
```

```python
# có thể sử dụng dtype để chỉ định kiểu dữ liệu mong muốn 
dtypes = {'POP': 'float32', 'AREA': 'float32', 'GDP': 'float32'}
df = pd.read_csv('newdata.csv', index_col=0, dtype=dtypes)
df.dtypes
```
